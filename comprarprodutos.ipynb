 Codigo de Menu de loja de produtos

```python
produtos_e_estoque = [
    ("Televisao", 5),
    ("Computador", 3),
    ("Geladeira", 8),
    ("Ar-Condicionado", 2),
    ("Iphone", 0),
]

opção = 0
while opção != 5:
    print(''' 
    [ 1 ] Ver produtos
    [ 2 ] Comprar
    [ 3 ] Vender
    [ 4 ] Cadastrar produto
    [ 5 ] Sair do programa''')
    
    opção = int(input("Escolha qual é a sua opção: "))

    if opção == 1:
        print("Produtos disponíveis:")
        for i, (produto, estoque) in enumerate(produtos_e_estoque):
            print(f"{i}: {produto} - Estoque: {estoque}")

    elif opção == 2:
        print("Opção de compra selecionada.")
        print("Produtos disponíveis:")
        for i, (produto, estoque) in enumerate(produtos_e_estoque):
            print(f"{i}: {produto} - Estoque: {estoque}")
        escolha = int(input("Qual produto você quer comprar? (Digite o número correspondente): "))
        if 0 <= escolha < len(produtos_e_estoque):
            produto, estoque = produtos_e_estoque[escolha]
            if estoque > 0:
                print(f"Você comprou um {produto}")
                produtos_e_estoque[escolha] = (produto, estoque - 1)
            else:
                print(f"{produto} está fora de estoque.")
        else:
            print("Opção inválida.")

    elif opção == 3:
        print("Opção de venda selecionada.")
        nome_produto = input("Digite o nome do produto: ")
        for i, (produto, estoque) in enumerate(produtos_e_estoque):
            if produto == nome_produto:
                quantidade_venda = int(input(f"Digite a quantidade de {produto} para vender: "))
                if quantidade_venda <= estoque:
                    produtos_e_estoque[i] = (produto, estoque - quantidade_venda)
                    print(f"{quantidade_venda} unidades de {produto} vendidas.")
                else:
                    print("Quantidade insuficiente em estoque.")
                break
        else:
            print("Produto não encontrado na lista.")

    elif opção == 4:
        nome_produto = input("Digite o nome do produto: ")
        quantidade_produto = int(input("Digite a quantidade em estoque: "))
        for i, (produto, estoque) in enumerate(produtos_e_estoque):
            if produto == nome_produto:
                produtos_e_estoque[i] = (produto, estoque + quantidade_produto)
                print(f"Estoque de {produto} atualizado para {estoque + quantidade_produto}.")
                break
        else:
            produtos_e_estoque.append((nome_produto, quantidade_produto))
            print(f"Produto '{nome_produto}' cadastrado com sucesso.")

    elif opção == 5:
        break  # Encerrar o loop quando a opção 5 for escolhida

    else: 
        print('Opção inválida. Tente novamente!')
    
    print('=-=' * 10)

print('Fim do programa! Volte sempre!')
```
